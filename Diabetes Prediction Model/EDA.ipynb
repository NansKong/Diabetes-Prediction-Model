```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the processed training set
df = pd.read_csv('data/processed/diabetes_train_processed.csv')
print(df.shape)
df.head()
```

```python
#visualisation

sns.countplot(x='Outcome', data=df)
plt.title('Diabetes Outcome Distribution')
plt.show()
```

```python
#histograms and boxplot visualise

df.hist(bins=20, figsize=(16,12))
plt.tight_layout()
plt.show()
```

```python
#boxplots
for col in df.columns:
    if col != 'Outcome':
        sns.boxplot(x='Outcome', y=col, data=df)
        plt.title(f'{col} by Diabetes Outcome')
        plt.show()
```

```python
#correlation matrix
corr = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```
